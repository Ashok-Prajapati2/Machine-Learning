```python
## replace outliers with median
## Columans = DurationOfPitch, NumberOfPersonVisiting, NumberOfFollowups, numberOfTrips,monthlyIncome
```

```python

def replace_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] > upper_bound, df[col].median(), df[col])
    return df
```

```python
for col in ['DurationOfPitch', 'NumberOfPersonVisiting', 'NumberOfFollowups', 'NumberOfTrips', 'MonthlyIncome']:
    df = replace_outliers(df, col)  
    
```

```python
categorical_features = [feature for feature in X.columns if X[feature].dtype == 'O']
num_columns = [feature for feature in X.columns if X[feature].dtype != 'O']
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
sclar = StandardScaler()
ohe = OneHotEncoder(drop='first')
preprocessor = ColumnTransformer(
    [
        ('num', sclar, num_columns),
        ('cat', ohe, categorical_features)
    ])
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import Perceptron

# For regression models
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
## hyperparameter tuning in random forest

## https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2'],  # Number of features to consider for the best split
    'bootstrap': [True, False],  # Whether bootstrap samples are used when building trees
}
grid_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
best_rf = grid_search.best_estimator_
y_pred_test = best_rf.predict(X_test)
accuracy_score(y_test, y_pred_test)*100
```

```python
models = {
    # Classification Models
    'LogisticRegression': LogisticRegression(),
    'SVC': SVC(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'GaussianNB': GaussianNB(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(bootstrap=False, max_depth=30, n_estimators=200),
    'GradientBoostingClassifier': GradientBoostingClassifier(),
    'AdaBoostClassifier': AdaBoostClassifier(),
    'Perceptron': Perceptron(),
    
    # Regression Models
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'SVR': SVR(),
    'DecisionTreeRegressor': DecisionTreeRegressor(),
    'RandomForestRegressor': RandomForestRegressor(),
    'GradientBoostingRegressor': GradientBoostingRegressor(),
}
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, f1_score, roc_auc_score, recall_score

model_reports = {}

for i in range(len(models)):
    model = list(models.values())[i]
    
    try:
        # Train the model
        model.fit(X_train, y_train)

        # Predict on train and test data
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)

        # Calculate metrics for test data
        accuracy_score_test = accuracy_score(y_test, y_pred_test)
        f1 = f1_score(y_test, y_pred_test, average='macro')
        precision_score_test = precision_score(y_test, y_pred_test, average='macro')
        recall_score_test = recall_score(y_test, y_pred_test, average='macro')

        # Calculate metrics for train data
        accuracy_train = accuracy_score(y_train, y_pred_train)
        f1_train = f1_score(y_train, y_pred_train, average='macro')
        precision_train = precision_score(y_train, y_pred_train, average='macro')
        recall_score_train = recall_score(y_train, y_pred_train, average='macro')

        # Save the results to the dictionary
        model_reports[model.__class__.__name__] = {
            'Test Accuracy': accuracy_score_test * 100,
            'Test F1': f1,
            'Test Precision': precision_score_test * 100,
            'Test Recall': recall_score_test * 100,
            'Train Accuracy': accuracy_train * 100,
            'Train F1': f1_train * 100,
            'Train Precision': precision_train * 100,
            'Train Recall': recall_score_train * 100,
        }

        # Print the results
        print(f"""
Model: {model.__class__.__name__}
Test Accuracy: {accuracy_score_test * 100:.3f}
Test F1: {f1:.3f}
Test Precision: {precision_score_test * 100:.3f}
Test Recall: {recall_score_test * 100:.3f}

Train Accuracy: {accuracy_train * 100:.3f}
Train F1: {f1_train * 100:.3f}
Train Precision: {precision_train * 100:.3f}
Train Recall: {recall_score_train * 100:.3f}
""")
    
    except Exception as e:
        print(f"Error occurred while training {model.__class__.__name__}: {e}")

```

```python
# save json
import json
with open('data.json', 'w') as f:
    json.dump(model_reports, f, indent=4)
```
