```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=100000, n_features=3, n_informative=2, n_redundant=1, n_classes=2
                          , random_state=999)
```

```python
X = pd.DataFrame(X, columns=['feature1', 'feature2', 'feature3'])
y = pd.Series(y, name='target')
```

```python
X.head()
```

```python
sns.pairplot(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10
                           , weights='uniform',algorithm='kd_tree', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)*100
```
