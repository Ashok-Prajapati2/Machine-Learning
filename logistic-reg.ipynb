```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
## create dataset
X,y = make_classification(n_samples=1000, n_features=10, n_classes=2, n_informative=5, n_redundant=5, random_state=1)
```

```python
X.shape, y.shape
```

```python
pd.DataFrame(X).head(), pd.DataFrame(y).head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,  classification_report, confusion_matrix, roc_auc_score    
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.accuracy = accuracy_score(y_test, y_pred)
lr.precision = precision_score(y_test, y_pred)
lr.recall = recall_score(y_test, y_pred)
lr.f1 = f1_score(y_test, y_pred)
lr.auc = roc_auc_score(y_test, y_pred)
lr.confusion_matrix = confusion_matrix(y_test, y_pred)
lr.classification_report = classification_report(y_test, y_pred)
```

```python
print('accuracy:', lr.accuracy*100)
print('precision:', lr.precision*100)
print('recall:', lr.recall*100)
print('f1:', lr.f1*100)
print('roc_auc:', lr.auc*100)
print('\n confusion_matrix:\n', lr.confusion_matrix)
print('\nclassification_report:\n', lr.classification_report)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(lr.confusion_matrix, annot=True)
```
